```python
import pandas as pd
```

```python
df=pd.read_csv('advertising.csv')
```

```python
df.head()
df.describe()
```

```python

# Check for missing values
print(df.isnull().sum())

# Check for data types and unique values
print(df.dtypes)
print(df.nunique())
```

```python
# Assuming there are no missing values, but if there are, handle them as follows:
df = df.fillna(df.mean())  # For numerical columns
```

```python
# Convert categorical variables to numerical values using one-hot encoding if necessary
df = pd.get_dummies(df, drop_first=True)
```

```python
from sklearn.preprocessing import StandardScaler

# Separate features and target variable
X = df.drop('Sales', axis=1)
y = df['Sales']

# Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)

# Initialize and train the models
model_lr = LinearRegression()
model_lr.fit(X_train, y_train)

model_rf = RandomForestRegressor()
model_rf.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Predictions
y_pred_lr = model_lr.predict(X_test)
y_pred_rf = model_rf.predict(X_test)

# Evaluate Linear Regression
print("Linear Regression Performance:")
print(f"Mean Absolute Error: {mean_absolute_error(y_test, y_pred_lr)}")
print(f"Mean Squared Error: {mean_squared_error(y_test, y_pred_lr)}")
print(f"R^2 Score: {r2_score(y_test, y_pred_lr)}")

# Evaluate Random Forest
print("\nRandom Forest Performance:")
print(f"Mean Absolute Error: {mean_absolute_error(y_test, y_pred_rf)}")
print(f"Mean Squared Error: {mean_squared_error(y_test, y_pred_rf)}")
print(f"R^2 Score: {r2_score(y_test, y_pred_rf)}")
```

```python
from sklearn.model_selection import GridSearchCV

# Define parameter grid for Random Forest
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_features': ['auto', 'sqrt', 'log2'],
    'max_depth': [None, 10, 20, 30],
}

# Initialize and perform grid search
grid_search = GridSearchCV(estimator=RandomForestRegressor(), param_grid=param_grid, cv=3, n_jobs=-1, verbose=2)
grid_search.fit(X_train, y_train)

# Best parameters and score
print("Best Parameters:", grid_search.best_params_)
print("Best Score:", grid_search.best_score_)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Get feature importances
importances = model_rf.feature_importances_
indices = np.argsort(importances)[::-1]

# Plot feature importances
plt.figure(figsize=(12, 8))
plt.title('Feature Importances')
plt.bar(range(X.shape[1]), importances[indices], align='center')
plt.xticks(range(X.shape[1]), [X.columns[i] for i in indices], rotation=90)
plt.xlabel('Feature')
plt.ylabel('Importance')
plt.show()
```
